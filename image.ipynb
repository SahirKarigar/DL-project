```python
from PIL import Image, ImageTk
from tkinter import filedialog, Label, Button, Tk, StringVar
import tkinter as tk
import numpy as np
from tensorflow.keras.models import load_model

# Load your model
model = load_model('classifier5.h5')

# Create the main window
my_w = Tk()
my_w.title("Image Classifier")

# Global variables for storing the resized image and PhotoImage
img_resized = None
photo = None

# Function to upload an image and make predictions
def upload_and_predict():
    global img_resized, photo

    # Upload image
    f_types = [('Image Files', '*.jpg;*.png')]
    filename = filedialog.askopenfilename(filetypes=f_types)
    img = Image.open(filename)

    # Convert the image to RGB
    img = img.convert('RGB')

    # Resize the image
    img_resized = img.resize((128, 128), resample=Image.LANCZOS)  # new width & height

    # Convert the resized image to a NumPy array
    img_array = np.array(img_resized)

    # Normalize the pixel values
    img_array = img_array.astype('float') / 255.0

    # Ensure the array has the correct shape (remove the alpha channel)
    img_array = img_array[:, :, :3]  # Keep only the first three channels (RGB)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

    # Make predictions
    prediction = model.predict(img_array)

    # Display the image on the GUI
    photo = ImageTk.PhotoImage(img_resized)
    image_label.config(image=photo)
    image_label.image = photo

    # Display the prediction result on the GUI
    result_text = f"Prediction: {'AI-generated portrait' if prediction > 0.5 else 'Real portrait'}"
    prediction_label_var.set(result_text)  # Update the StringVar

# Widgets
upload_and_predict_button = Button(my_w, text='Upload and Predict', command=upload_and_predict)
upload_and_predict_button.grid(row=1, column=0, columnspan=2, pady=10)

image_label = Label(my_w)
image_label.grid(row=2, column=0, columnspan=2, pady=10)

prediction_label_var = StringVar()
prediction_label = Label(my_w, textvariable=prediction_label_var, font=('Helvetica', 12, 'bold'))
prediction_label.grid(row=3, column=0, columnspan=2, pady=10)

# Start the main loop
my_w.mainloop()
```
